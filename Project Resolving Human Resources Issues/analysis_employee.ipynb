```python
# Memilih data numerik
# Menggunakan pandas untuk memilih kolom numerik
numeric_data = data.select_dtypes(include=[np.number])

# Menghitung korelasi
correlation = numeric_data.corr()

# Menampilkan korelasi
correlation
```

```python
# Memeriksa nilai yang hilang
missing_values = data.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Memeriksa data duplikat
duplicate_data = data.duplicated().sum()
duplicate_data
```

```python
# Memeriksa inkonsistensi data
for col in data.columns:
    if data[col].dtype == 'object':
        unique_values = data[col].unique()
        print(f"Unique values in {col}: {unique_values}")
```

```python
# Memeriksa outlier
# Menggunakan boxplot untuk memeriksa outlier
plt.figure(figsize=(16, 8))
sns.boxplot(data=data[numeric_data.columns])
plt.title('Boxplot of Numeric Data')
plt.xticks(rotation=45)
plt.show()

# Menghitung outlier
outliers = {}
for col in numeric_data.columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers[col] = data[(data[col] < lower_bound) | (data[col] > upper_bound)].shape[0]
outliers
```
